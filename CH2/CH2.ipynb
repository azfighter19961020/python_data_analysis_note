## 實訓一 創建數組並進行運算

1. 創建一個數值範圍為0~1,間隔為0.01的數組
2. 創建100個服從正態分布的隨機數
3. 對創建的兩個數組進行四則運算
4. 對創建的隨機數組進行簡單的統計分析

```python
#導入numpy模塊
import numpy as np
```

```python
#創建數值範圍為0~1,間隔為0.01的數組
arr1=np.arange(0,1,0.01)
print('創建的數組1為:\n',arr1)
```

```python
#創建100個服從正態分布的隨機數
arr2=np.random.randn(100)
print('創建的數組2為:\n',arr2)
```

```python
#對創建的兩個數組進行四則運算
print('兩者相加為:\n',arr1+arr2)
```

```python
print('兩者相減為:\n',arr1-arr2)
```

```python
print('兩者相乘為:\n',arr1*arr2)
```

```python
print('兩者相除為:\n',arr1/arr2)
```

```python
print('兩者進行冪運算為:\n',arr1 ** arr2)
```

```python
#對隨機數組arr2進行統計分析
print('數組arr2的和為:',np.sum(arr2))
```

```python
print('數組arr2的均值為:',arr2.mean())
```

```python
print('數組arr2的標準差為:',arr2.std())
```

```python
print('數組arr2的方差為:',arr2.var())
```

```python
print('數組arr2的最小值為:',np.min(arr2))
```

```python
print('數組arr2的最大值為:',np.max(arr2))
```

```python
print('數組arr2的最小值索引為:',np.argmin(arr2))
```

```python
print('數組arr2的最大值索引為:',np.argmax(arr2))
```

```python
#沿著縱軸求和,數組arr2為一維數組,故只有縱軸
print('數組arr2沿著縱軸求和為:',arr2.sum(axis=0))
```

```python
#非聚合計算
print('數組arr2的累計和為:\n',np.cumsum(arr2))
```

```python
print('數組arr2的累計積為:\n',np.cumprod(arr2))
```

## 實訓二 創建一個國際象棋的棋盤

>創建國際棋盤,填充8x8矩陣

1. 創建一個8x8的矩陣
2. 把1.3.5.7列和2.4.6行的元素設為一

```python
#創建8x8矩陣
chess=np.zeros((8,8))
print('創建矩陣為:\n',chess)
```

```python
for i in range(0,7,2):
    for j in range(0,7,2):
        chess[i][j]=1
else:
    for i in range(1,8,2):
        for j in range(1,8,2):
            chess[i][j]=1
            
print('創建的棋盤為:\n',chess)
```
