## 實訓一 使用sklearn處理wine和wine_quality數據集

- 需求說明
>
>wine數據集包含三種不同起源的葡萄酒的紀錄,共178條
>
>其中,每個特徵對應葡萄酒的每個化學成分,並且都屬於連續型數據
>
>通過化學分析可推斷葡萄酒的來源
>
>wine_quality數據集共有4898個觀察值,11個輸入特徵和一個標籤
>
>其中,不同類的觀察值數量不等,所有特徵為連續型數據
>
>通過酒的各類化學成分,預測該葡萄酒的評分
>

- 實現步驟
  1. 使用pandas庫分別讀取wine數據集與wine_quality數據集
  2. 將wine數據集與wine_quality數據集的數據與標籤拆分開
  3. 將wine_quality數據集劃分為訓練集與測試集
  4. 標準化wine數據集和wine_quality數據集
  5. 對wine數據集和wine_quality數據集進行PCA降維

```python
#導入模塊
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
```

```python
#導入數據
Wine=pd.read_csv('./wine.csv',encoding='gbk')
WineQuality=pd.read_csv('./winequality.csv',sep=';',encoding='gbk')
```

```python
print(Wine.head())
```

```python
print(WineQuality.head())
```

```python
#取出Wine標籤及資料
Wine_data=Wine.iloc[:,0:]
Wine_target=Wine.iloc[:,0]
print('Wine資料的形狀為:',Wine_data.shape)
print('Wine標籤的形狀為:',Wine_target.shape)
```

```python
#取出WineQuality標籤及資料
WineQuality_data=WineQuality.iloc[:,0:10]
WineQuality_target=WineQuality.iloc[:,10]
print('WineQuality資料的形狀為:',WineQuality_data.shape)
print('WineQuality標籤的形狀為:',WineQuality_target.shape)
```

```python
#將WineQuality劃分為訓練集與測試集
WineQualityDataTrain,WineQualityDataTest, \
WineQualityTargetTrain,WineQualityTargetTest= \
train_test_split(WineQuality_data,WineQuality_target, \
                test_size=0.2,random_state=42)
```

```python
print('WineQuality訓練集數據的形狀為:',WineQualityDataTrain.shape)
print('WineQuality訓練集標籤的形狀為:',WineQualityTargetTrain.shape)
print('WineQuality測試集數據的形狀為:',WineQualityDataTest.shape)
print('WineQuality測試集標籤的形狀為:',WineQualityTargetTest.shape)
```

```python
#標準化Wine與WineQuality數據集
ScalerWine=MinMaxScaler().fit(Wine)
WineScaler=ScalerWine.transform(Wine)
ScalerWineQuality=MinMaxScaler().fit(WineQuality)
WineQualityScaler=ScalerWineQuality.transform(WineQuality)
```

```python
#對Wine與WineQuality進行PCA降維
PCAmodel_Wine=PCA(n_components=10).fit(WineScaler)
WinePCA=PCAmodel_Wine.transform(WineScaler)
PCAmodel_WineQuality=PCA(n_components=10).fit(WineQualityScaler)
WineQualityPCA=PCAmodel_WineQuality.transform(WineQualityScaler)
```

```python
print('PCA降維前WineScaler的形狀為:',WineScaler.shape)
print('PCA降維後WineScaler的形狀為:',WinePCA.shape)
print('PCA降維前WineQualityScaler的形狀為:',WineQualityScaler.shape)
print('PCA降維後WineQualityScaler的形狀為:',WineQualityPCA.shape)
```

## 實訓二 構建基於Wine數據集的K-Means聚類模型

- 需求說明
>
>Wine數據集的葡萄酒總共分為三類,通過將Wine數據集的數據進行聚類,聚集為3個簇,實現類別劃分
>

- 實現步驟
  1. 根據實訓一的Wine數據集處理的結果,構建聚類數目為3的K-Means模型
  2. 對比真實標籤和聚類標籤求取FMI
  3. 在聚類數目為2~10類時,確定最優聚類數目
  4. 求取模型的輪廓係數,繪製輪廓係數折線圖,確定最優聚類數目
  5. 求取Calinski-Harabasz指數,確定最優聚類數目

```python
#導入模塊
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
#創建聚類數目為3的模型
Winekmeans=KMeans(n_clusters=3,random_state=123).fit(WineScaler)
print('創建Wine的數目為3的K-Means模型為:\n',Winekmeans)
```

```python
#聚類數目為3的結果可視化

#使用TSNE降維,將數據降成2維
tsne=TSNE(n_components=2,init='random',random_state=177).fit(WineScaler)
df=pd.DataFrame(tsne.embedding_)
df['labels']=Winekmeans.labels_
#提取不同標籤的數據
df1=df[df['labels']==0]
df2=df[df['labels']==1]
df3=df[df['labels']==2]
#繪製圖形
p1=plt.figure(figsize=(9,6))
plt.plot(df1[0],df1[1],'bo',df2[0],df2[1],'r*',df3[0],df3[1],'gD')
plt.show()
```

```python
#導入模塊
from sklearn.metrics import fowlkes_mallows_score
#查看數目為3的FMI評價
score3=fowlkes_mallows_score(Wine_target,Winekmeans.labels_)
print('Wine數據集聚類為3的FMI評價為:',score3)
```

```python
#在聚類為2~10時求最優聚類數目
for i in range(2,11):
    kmeans=KMeans(n_clusters=i,random_state=123).fit(WineScaler)
    score=fowlkes_mallows_score(Wine_target,kmeans.labels_)
    print('Wine數據集聚類為%d的FMI評價為:%f' % (i,score))
```

```python
#使用輪廓係數法評價2~10聚類的模型
from sklearn.metrics import silhouette_score
silhouetteScore=[] #宣告空陣列存放結果
for i in range(2,11):
    kmeans=KMeans(n_clusters=i,random_state=123).fit(WineScaler)
    score=silhouette_score(WineScaler,kmeans.labels_)
    silhouetteScore.append(score)
    
plt.figure(figsize=(9,6))
plt.plot(range(2,11),silhouetteScore,linewidth=1.5,linestyle='-')
plt.show()
```

```python
#使用Calinski-Harabasz評價
from sklearn.metrics import calinski_harabaz_score
for i in range(2,11):
    kmeans=KMeans(n_clusters=i,random_state=123).fit(WineScaler)
    score=calinski_harabaz_score(WineScaler,kmeans.labels_)
    print('Wine數據集聚類為%d的calinski_harabaz指數為:%f' % (i,score))
```

>
>綜合FMI評價.輪廓係數與calinski-harabasz評價結果
>
>在聚類數目為3時模型最佳

## 實訓三 構建基於wine數據集的SVM分類模型

- 需求說明
>
>Wine數據集中的葡萄酒類別為三種,將Wine數據集分為訓練集和測試集,
>
>使用訓練集訓練SVM分類模型,並使用訓練完成的模型預測測試集的葡萄酒類別歸屬
>

- 實現步驟
  1. 讀取Wine數據集,區分標籤和數據
  2. 將Wine數據集劃分為訓練集與測試集
  3. 使用離差標準化方法標準化wine數據集
  4. 構建SVM模型,並預測測試集結果
  5. 打印出分類報告,評價分類模型性能

```python
#取出Wine標籤與資料
WineData=Wine.iloc[:,0:]
WineTarget=Wine.iloc[:,0]
```

```python
#劃分測試集與訓練集
WineDataTrain,WineDataTest, \
WineTargetTrain,WineTargetTest= \
train_test_split(WineData,WineTarget, \
                test_size=0.2,random_state=22)
```

```python
#離差標準化數據集
#生成規則
WineDataScaler=MinMaxScaler().fit(WineDataTrain)
#應用於測試集與訓練集
WineDataTrainScaler=WineDataScaler.transform(WineDataTrain)
WineDataTestScaler=WineDataScaler.transform(WineDataTest)
```

```python
#創建SVM模型
from sklearn.svm import SVC
WineSVM=SVC().fit(WineDataTrainScaler,WineTargetTrain)
print('創建的SVM模型為:\n',WineSVM)
```

```python
#預測測試集結果
WineTargetPredict=WineSVM.predict(WineDataTestScaler)
#打印出分類報告
import numpy as np
true=np.sum(WineTargetPredict==WineTargetTest)
print('預測對的結果數目為:',true)
print('預測錯的結果數目為:',WineTargetTest.shape[0]-true)
print('預測結果準確率為:',true/WineTargetTest.shape[0])
```

## 實訓三 構建基於wine_quality數據集的回歸模型

- 需求說明
>
>wine_quality數據集的葡萄酒評分在1~10之間,
>
>構建線性回歸模型與梯度提升回歸模型
>
>訓練wine_quality數據集的訓練集數據,訓練完成後預測測試集的葡萄酒評分
>
>結合真實評分,評價購建的兩個回歸模型的好壞
>

- 實現步驟
  1. 根據Wine_quality數據集處理的結果,構建線性回歸模型
  2. 根據Wine_quality數據集處理的結果,構建梯度提升回歸模型
  3. 結合真實評分與預測評分,計算均方誤差.中值絕對誤差.可解釋方差值
  4. 根據得分,判定模型的性能優劣

```python
#構建線性回歸模型
from sklearn.linear_model import LinearRegression
clf=LinearRegression().fit(WineQualityDataTrain,WineQualityTargetTrain)
print('建立的線性回歸模型為:\n',clf)
```

```python
#構建梯度提升回歸模型
from sklearn.ensemble import GradientBoostingRegressor
GBR=GradientBoostingRegressor().fit(WineQualityDataTrain,WineQualityTargetTrain)
print('建立的梯度提升回歸模型為:\n',GBR)
```

```python
#使用線性回歸模型預測測試集
WineQualityclf=clf.predict(WineQualityDataTest)
```

```python
#線性回歸結果可視化
plt.rcParams['font.sans-serif']='SimHei'
p2=plt.figure(figsize=(15,14))
plt.plot(range(WineQualityDataTest.shape[0]),WineQualityTargetTest,color='blue',linewidth=1.5,linestyle='-')
plt.plot(range(WineQualityDataTest.shape[0]),WineQualityclf,color='red',linewidth=1.5,linestyle='-.')
plt.legend(['真實值','預測值'])
plt.show()
```

```python
#使用梯度提升回歸模型預測
WineQualityGBR=GBR.predict(WineQualityDataTest)
```

```python
#梯度提升回歸結果可視化
plt.rcParams['font.sans-serif']='SimHei'
p3=plt.figure(figsize=(15,14))
plt.plot(range(WineQualityDataTest.shape[0]),WineQualityTargetTest,color='blue',linewidth=1.5,linestyle='-')
plt.plot(range(WineQualityDataTest.shape[0]),WineQualityGBR,color='red',linewidth=1.5,linestyle='-.')
plt.legend(['真實值','預測值'])
plt.show()
```

```python
#導入模塊
from sklearn.metrics import mean_squared_error, \
median_absolute_error,explained_variance_score
```

```python
#評斷線性回歸模型
print('WineQuality線性回歸模型的均方誤差為:',mean_squared_error(WineQualityTargetTest,WineQualityclf))
print('WineQuality線性回歸模型的中值絕對誤差為:',median_absolute_error(WineQualityTargetTest,WineQualityclf))
print('WineQuality線性回歸模型的可解釋方差值為:',explained_variance_score(WineQualityTargetTest,WineQualityclf))
```

```python
#評斷梯度提升回歸模型
print('WineQuality梯度提升回歸模型的均方誤差為:',mean_squared_error(WineQualityTargetTest,WineQualityGBR))
print('WineQuality梯度提升回歸模型的中值絕對誤差為:',median_absolute_error(WineQualityTargetTest,WineQualityGBR))
print('WineQuality梯度提升回歸模型的可解釋方差值為:',explained_variance_score(WineQualityTargetTest,WineQualityGBR))
```
