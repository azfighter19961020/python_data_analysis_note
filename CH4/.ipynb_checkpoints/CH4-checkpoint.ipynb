## 實訓二 提取用戶信息更新表和登錄信息表的時間信息

- 需求說明
>
>用戶信息更新表和登錄信息表存在大量時間數據,提取時間數據內存在的信息
>
>用戶登錄時間.借款成交時間.用戶信息更新時間的時間差信息能夠反映出P2P網路貸款不同用戶的行為信息
>

- 實現步驟
  1. 使用to_datetime函數轉換用戶信息更新表和登錄信息表的時間字符串
  2. 使用year.month.week等方法提取用戶信息更新表和登錄訊息表的時間信息
  3. 計算用戶信息更新表和登錄信息表中兩時間的差,分別以日.小時.分鐘計算

```python
#導入數據
loginfo=pd.read_csv('./Training_LogInfo.csv',encoding='gbk') #登錄信息表
userupdate=pd.read_csv('./Training_UserUpdate.csv',encoding='gbk')
```

```python
print('loginfo_columns:\n',loginfo.columns)
print('userupdateinfo_columns:\n',userupdate.columns)
```

```python
#轉換時間字符串,loginfo的時間在LogInfo3,userupdate的時間在UserupdateInfo2
print('轉換前LogInfo3的類型為:',loginfo['LogInfo3'].dtypes)
loginfo['LogInfo3']=pd.to_datetime(loginfo['LogInfo3'])
print('轉換後LogInfo3的類型為:',loginfo['LogInfo3'].dtypes)
```

```python
print('轉換前UserupdateInfo2的類型為:',userupdate['UserupdateInfo2'].dtypes)
userupdate['UserupdateInfo2']=pd.to_datetime(userupdate['UserupdateInfo2'])
print('轉換後UserupdateInfo2的類型為:',userupdate['UserupdateInfo2'].dtypes)
```

```python
#提取year.month.weekday_name.day-登錄信息表
LogYear=[i.year for i in loginfo['LogInfo3']]
LogMonth=[i.month for i in loginfo['LogInfo3']]
LogWeeklyName=[i.weekday_name for i in loginfo['LogInfo3']]
LogDay=[i.day for i in loginfo['LogInfo3']]
```

```python
print('登錄信息表前五個年數據為:\n',LogYear[:5])
print('登錄信息表前五個月數據為:\n',LogMonth[:5])
print('登錄信息表前五個星期數據為:\n',LogWeeklyName[:5])
print('登錄信息表前五個日數據為:\n',LogDay[:5])
```

```python
#提取year.month.weekday_name.day-用戶信息更新表
UserYear=[i.year for i in userupdate['UserupdateInfo2']]
UserMonth=[i.month for i in userupdate['UserupdateInfo2']]
UserWeeklyName=[i.weekday_name for i in userupdate['UserupdateInfo2']]
UserDay=[i.day for i in userupdate['UserupdateInfo2']]
```

```python
print('用戶信息更新表前五個年數據為:\n',UserYear[:5])
print('用戶信息更新表前五個月數據為:\n',UserMonth[:5])
print('用戶信息更新表前五個星期數據為:\n',UserWeeklyName[:5])
print('用戶信息更新表前五個日數據為:\n',UserDay[:5])
```

```python
#求時間差
timedelta=loginfo['LogInfo3']-userupdate['UserupdateInfo2']
print('前後時間差的前五個為:\n',timedelta[:5])
```

## 實訓三 使用分組聚合方法分析兩表

- 需求說明
>
>分析兩表時,除了提取時間信息外,還可結合用戶編號進行分組聚合,進行組內分析
>
>通過組內分析可得出每組組內最早和最晚的信息更新時間.登錄時間.信息更新的次數.登錄次數等訊息
>

- 實現步驟
  1. 使用groupby方法對用戶信息更新表和登錄信息表進行分組
  2. 使用agg方法取分組後最早和最晚更新及登錄時間
  3. 使用size方法取分組後數據的更新次數與登錄次數

```python
#分組數據
LogGroup=loginfo.groupby('Idx')
UserUpdateGroup=userupdate.groupby('Idx')
print(LogGroup)
print(UserUpdateGroup)
```

```python
#導入numpy數據庫
import numpy as np
```

```python
#取登錄信息最晚登錄時間
print('登錄信息表各用戶最晚登錄時間為:\n',LogGroup['LogInfo3'].agg(np.max))
```

```python
#取登錄信息最早登錄時間
print('登錄信息表各用戶最早登陸時間為:\n',LogGroup['LogInfo3'].agg(np.min))
```

```python
#取用戶信息更新表最晚時間
print('用戶信息更新表各用戶最晚更新時間為:\n',UserUpdateGroup['UserupdateInfo2'].agg(np.max))
```

```python
#取用戶信息更新表最早時間
print('用戶信息更新表各用戶最早更新時間為:\n',UserUpdateGroup['UserupdateInfo2'].agg(np.min))
```

```python
#取登錄信息表的登錄次數
print('登錄信息表各用戶的登錄次數為:\n',LogGroup.size())
```

```python
#取用戶信息更新表的更新次數
print('用戶信息更新表各用戶的更新次數為:\n',UserUpdateGroup.size())
```

## 實訓四 對兩表進行長寬表轉換

- 需求說明
>
>用戶更新表與登錄信息表為長表,但主表為寬表,須通過長寬表轉換
>
>將數據合併在一張以用戶編號為主鍵的表內
>

- 實現步驟
  1. 使用pivot_table進行長寬表轉換
  2. 使用crosstab進行長寬表轉換

```python
#使用pivot_table轉換登錄信息表
LogPivot=pd.pivot_table(loginfo,index=['Idx','LogInfo3'],aggfunc=np.sum)
print('轉換後的登錄信息表為:\n',LogPivot)
```

```python
#使用pivot_table轉換用戶信息更新表
UserUpdatePivot=pd.pivot_table(userupdate,index=['Idx','UserupdateInfo2'],aggfunc=np.sum)
print('轉換後的用戶信息更新表為:\n',UserUpdatePivot)
```

```python
#使用crosstab轉換登錄信息表
LogCross=pd.crosstab(index=loginfo['Idx'],columns=loginfo['LogInfo3'])
```

```python
#使用crosstab轉換用戶信息更新表
UserUpdateCross=pd.crosstab(index=userupdate['Idx'],columns=userupdate['UserupdateInfo2'])
```

```python
print('使用crosstab轉換的登錄信息表為:\n',LogCross)
```

```python
print('使用crosstab轉換的用戶更新信息表為:\n',UserUpdateCross)
```
